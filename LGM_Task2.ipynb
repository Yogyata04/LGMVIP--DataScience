Importing all requride libraries

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Loading Dataset Of Terrorism

```python
data=pd.read_csv("/content/drive/MyDrive/globalterrorismdb_0718dist.csv", encoding="latin1")
df=pd.DataFrame(data)
print("Data has been successfully imported")
df
```

```python
df.info()
```

```python
df.columns
```

```python
# Printing all the Feature of Dataset
for i in df.columns:
    print(i, end=", ")
    print()

  
```

Cleaning The Dataset

```python
df=df[["iyear","imonth","iday","country_txt","region_txt","provstate","city",
       "latitude","longitude","location","summary","attacktype1_txt","targtype1_txt",
       "gname","motive","weaptype1_txt","nkill","nwound","addnotes"]]
df.head()
```

```python
# Renaming the Columns 
df.rename(columns={"iyear":"Year","imonth":"Month","iday":"Day","country_txt":"Country",
                   "region_txt":"Region","provstate":"Province/State","city":"City",
                   "latitude":"Latitude","longitude":"Longitude","location":"Location",
                   "summary":"Summary","attacktype1_txt":"Attack Type","targtype1_txt":"Target Type",
                   "gname":"Group Name","motive":"Motive","weaptype1_txt":"Weapon Type",
                   "nkill":"Killed","nwound":"Wounded","addnotes":"Add Notes"},inplace=True)

df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.shape
```

```python
# Checking the Null value in the data

df.isnull().sum()
```

```python
# Filling the null value in the dataset
df["Killed"]=df["Killed"].fillna(0)
df["Wounded"]=df["Wounded"].fillna(0)
df["Casualty"]=df["Killed"]+df["Wounded"]
```

```python
df.describe()
```

# **Visualizing the data**
# 1. Year wise Attacks
* Number of Attacks in each Year

```python
attacks=df["Year"].value_counts(dropna=False).sort_index().to_frame().reset_index().rename(columns={"index":"Year","Year":"Attacks"}).set_index("Year")
attacks.head(10)
```

```python
attacks.plot(kind="bar",color="red",figsize=(15,6),fontsize=13)
plt.title("Timeline of Attacks",fontsize=15)
plt.xlabel("Years",fontsize=15)
plt.ylabel("Number of Attacks",fontsize=15)
plt.show()
```

From the above graph , its concluded that mostly attacks happen in 2014,2015,2016,2017.
1.  Most number of attacks(16903) in 2014
2. Least number of attacks(471) in 1971



*   **Total Casualties (Killed + Wounded) in each Year**



```python
TC=df[["Year","Casualty"]].groupby("Year").sum()
TC.head()
```

```python
TC.plot(kind="bar",color="red",figsize=(15,6))
plt.title("Year wise Casualties",fontsize=13)
plt.xlabel("Years",fontsize=13)
plt.xticks(fontsize=12)
plt.ylabel("Number of Casualties",fontsize=13)
plt.show()
```

1. More number of the casualties in the year 2014 which is nearly 8000+.
2. Least number of the casualties in the year 1971.

```python
# Killed people in each year
yk=df[["Year","Killed"]].groupby("Year").sum()
yk.head()
```

```python
# Wounded in each Region
yw=df[["Year","Wounded"]].groupby("Year").sum()
yw.head()
```

```python
fig=plt.figure()
ax0=fig.add_subplot(2,1,1)
ax1=fig.add_subplot(2,1,2)

#Killed
yk.plot(kind="bar",color="red",figsize=(15,15),ax=ax0)
ax0.set_title("People Killed in each Year")
ax0.set_xlabel("Years")
ax0.set_ylabel("Number of People Killed") 

#Wounded
yw.plot(kind="bar",color="red",figsize=(15,15),ax=ax1)
ax1.set_title("People Wounded in each Year")
ax1.set_xlabel("Years")
ax1.set_ylabel("Number of People Wounded")

plt.show()
```

# 2. Region wise Attacks
1. Distribution of Terrorist Attacks over Regions from 1970-2017

```python
region =pd.crosstab(df.Year,df.Region)
region.head()
```

```python
region.tail()
```

```python
region.plot(kind="area", stacked=False, alpha=0.5,figsize=(20,10))
plt.title("Region wise attacks",fontsize=20)
plt.xlabel("Years",fontsize=20)
plt.ylabel("Number of Attacks",fontsize=20)
plt.show()
```

2. Total Terrorist Attacks in each Region from 1970-2017

```python
regiont=region.transpose()
regiont["Total"]=regiont.sum(axis=1)
ra=regiont["Total"].sort_values(ascending=False)
ra
```

```python
ra.plot(kind="bar",figsize=(15,6),color="green")
plt.title("Total Number of Attacks in each Region from 1970-2017")
plt.xlabel("Region")
plt.ylabel("Number of Attacks")
plt.show()
```

* Total Casualties (Killed + Wounded) in each Region

```python
rc=df[["Region","Casualty"]].groupby("Region").sum().sort_values(by="Casualty",ascending=False)
rc
```

```python
rc.plot(kind="bar",color="green",figsize=(14,5))
plt.title("Region wise Casualties",fontsize=13)
plt.xlabel("Regions",fontsize=13)
plt.xticks(fontsize=12)
plt.ylabel("Number of Casualties",fontsize=13)
plt.show()
```



*   No. of people killed in each region by terrorist



```python
r_Killed=df[["Region","Killed"]].groupby("Region").sum().sort_values(by="Killed",ascending=False)
r_Killed
```

* No. of People wounded in each region

```python
rw=df[["Region","Wounded"]].groupby("Region").sum().sort_values(by="Wounded",ascending=False)
rw
```

```python
fig=plt.figure()
ax0=fig.add_subplot(1,2,1)
ax1=fig.add_subplot(1,2,2)

#Killed
r_Killed.plot(kind="bar",ax=ax0,figsize=(20,10),color="green")
ax0.set_title("People Killed in each Region")
ax0.set_xlabel("Regions")
ax0.set_ylabel("Number of People Killed")

#Wounded
rw.plot(kind="bar",figsize=(15,6),ax=ax1,color="green")
ax1.set_title("People Wounded in each Region")
ax1.set_xlabel("Regions")
ax1.set_ylabel("Number of People Wounded")
```
